```python
# 4.创建一个DateFrame(df)，用data做数据，label做行索引
import numpy as np
data = {
    "animal": ["cat","cat","snake","dog","dog","cat","snake","cat","dog","dog"],
    "age":[2.5,3,0.5,np.nan,5,2,4.5,np.nan,7,3],
    "visits":[1,3,2,3,2,3,1,1,2,1],
    "priority":["yes","yes","no","yes","no","no","no","yes","no","no"]
}

label = ["a","b","c","d","e","f","g","h","i","j"]
```

```python
df = pd.DataFrame(data=data,index=label)
df.head()
```

```python
# 5.显示有关此df及其数据的基本信息的摘要
df.describe()
# or
df.info()
```

```python
# 6.查看此df的前三行
df.head(3)
```

```python
# 7.选择df中列标签为animal和age的数据
df.loc[:,["animal","age"]]
```

```python
# 8.选择行为[3,4,8]，且列为["animal","age"]
df.iloc[[3,4,8]].loc[:,["animal","age"]]
# or
df.loc[df.index[[3,4,8]],["animal","age"]]
```

```python
# 9.选择visits大于3的行
df[df["visits"]>3]
```

```python
# 10.选择age为缺失值的行
df[df["age"].isnull()]
# or
df[df["age"].isna()]
```

```python
# 11.选择animal为cat，且age小于3的行
df[(df.animal=="cat") & (df.age<3)]
```

```python
# 12.选择age在2到4之间的数据(包含边界值)
df[df.age.between(2,4,inclusive=True)]
```

```python
# 13.将f行的age改为1.5
df.loc["f","age"]=1.5
df.head(10)
```

```python
# 14.计算visits列的数据总和
df.visits.sum()
```

```python
# 15.计算每种animal的平均age，visits
df.groupby("animal")[["age","visits"]].mean()
```

```python
# 16.追加一行(k)，列的数据自定义，然后再删除新追加的k行
df.loc["k"] = df.loc["a"].values
df.drop("k", inplace=True)
```

```python
# 17.计算每种animal的个数(cate有几个，dog有几个)
df.groupby("animal").size()
# or
df.animal.value_counts()
```

```python
# 18.先根据age降序，再根据visits升序排列
df.sort_values(by=["age","visits"],ascending=[False,True])
```

```python
# 19.将priority列的yes和no用True和False替换
df.priority = df.priority.replace(to_replace=["yes","no"],value=[True,False])
# or map方法
df.priority = df.priority.map({"yes":True,"no":False})
```

```python
df.head()
```

```python
# 20.将animal列的snake用python替换
df.animal = df.animal.replace(to_replace="snake",value="python")
df.head()
```

```python
# 21.对于每种动物类型和每种访问次数，求出平均年龄。换句话说，每一行都是动物，每一列都是访问次数，其值是平均年龄(提示：使用数据透视表)
df.pivot_table(index="animal", columns="visits", values="age", aggfunc="mean")
```
