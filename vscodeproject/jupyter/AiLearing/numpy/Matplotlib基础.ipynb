```python
t = linspace(0, 2*pi, 50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
```

```python
subplot(1, 2, 1)
plot(x)
subplot(1, 2, 2)
plot(y)
```

```python
plot(x)
hold(False)
plot(y)
# 
hold(True)
```
